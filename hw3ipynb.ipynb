```python
import numpy as np
import matplotlib.pyplot as plt

# params
c = 3e8
lam = 1064e-9
rx = 1.0
Lx = 1000.0
Ly = 1.0

# carrier params
omega0 = 2 * np.pi * c / lam
k0 = 2 * np.pi / lam

# phases
phi_x = k0 * Lx
phi_y = k0 * Ly
phi_c = 0.5 * (phi_x + phi_y)
phi_d = 0.5 * (phi_x - phi_y)

f = np.logspace(0, 6, 5000)   # freq span
Omega = 2 * np.pi * f

P_I = -rx**2 * np.sin(Omega * phi_c / omega0) * np.sin(4 * phi_d) * np.sin(Omega * phi_d / omega0)
P_Q = -rx**2 * np.cos(Omega * phi_c / omega0) * np.sin(4 * phi_d) * np.sin(Omega * phi_d / omega0)
P = P_I + 1j * P_Q

mag = np.abs(P)
phase = np.angle(P, deg=True)

# bode plots
plt.figure(figsize=(8, 5))
plt.semilogx(f, mag, linewidth=2)
plt.xlabel('Frequency [Hz]')
plt.ylabel("P_as / gamma")
plt.title(r'Magnitude of P_as / gamma')
plt.grid(True, which='both', alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.semilogx(f, phase, linewidth=2)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [deg]')
plt.title(r'Phase of P_as / gamma')
plt.grid(True, which='both', alpha=0.3)
plt.tight_layout()
plt.show()
```
